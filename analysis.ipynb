```python
import requests
import json
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = input("Enter the URL of the page to scrape: ")
r = requests.get(url)
soup = bs(r.content)
json_fotmob = json.loads(soup.find('script', attrs={'id': '__NEXT_DATA__'}).contents[0])
```

```python
df = pd.DataFrame(json_fotmob['props']['pageProps']['content']['playerStats'])
df
```

```python
df = df.T
df.head()
```

```python
df['player_id'] = df.index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
new_df = df[['teamName', 'player_id']].copy()
new_df
```

Shots

```python
df_shots = pd.DataFrame(json_fotmob['props']['pageProps']['content']['shotmap']['shots'])
df_shots.head()
```

```python
new_df['player_id'] = new_df['player_id'].astype(int)
new_df.info()
```

```python
df_shots['teamName'] = df_shots['playerId'].map(new_df.set_index('player_id')['teamName'])
df_shots
```

```python
df_shots = pd.merge(
    df_shots,
    new_df[['player_id', 'teamName']],
    left_on='playerId',
    right_on='player_id',
    how='left'  # keeps all rows from df_shots even if no match
).drop('player_id', axis=1)  # drop the redundant column

df_shots
```

```python
# DF to CSV
df_shots.to_csv('/home/axel/Code/Python/axel/streamlit/csv/CerroBolivar.csv', index=False)
```
