```python
import requests
import json
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = input("Enter the URL of the page to scrape: ")
r = requests.get(url)
soup = bs(r.content)
json_fotmob = json.loads(soup.find('script', attrs={'id': '__NEXT_DATA__'}).contents[0])
```

```python
# General match info
json_fotmob['props']['pageProps']['general']
```

```python
matchRound = int(json_fotmob['props']['pageProps']['general']['matchRound'])
matchRound
```

```python
type(matchRound)
```

```python
h_team = json_fotmob['props']['pageProps']['general']['homeTeam']['name']
```

```python
json_fotmob['props']['pageProps']['general']['awayTeam']['name']
```

```python
df = pd.DataFrame(json_fotmob['props']['pageProps']['content']['playerStats'])
df
```

```python
df = df.T
df.head()
```

```python
df['player_id'] = df.index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
new_df = df[['teamName', 'player_id']].copy()
new_df
```

Shots

```python
df_shots = pd.DataFrame(json_fotmob['props']['pageProps']['content']['shotmap']['shots'])
df_shots.head()
```

```python
# Add match round to shots DF
df_shots['matchRound'] = matchRound
df_shots.head()
```

```python
new_df['player_id'] = new_df['player_id'].astype(int)
new_df.info()
```

```python
# Compare the two dataframes
df_shots['teamName'] = df_shots['playerId'].map(new_df.set_index('player_id')['teamName'])
df_shots.head()
```

```python
# Same result as above, but using merge
df_shots = pd.merge(
    df_shots,
    new_df[['player_id', 'teamName']],
    left_on='playerId',
    right_on='player_id',
    how='left'  # keeps all rows from df_shots even if no match
).drop('player_id', axis=1)  # drop the redundant column

df_shots
```

Options to get home or away team

```python
# Get home and away team -> worked fine until I hit Club Atletico Penarol vs Peñarol | Check DeepSick
# df_shots['h_a'] = df_shots['teamName'].apply(lambda x: 'h' if x == h_team else 'a') # does not work with Botafogo RJ & Botafogo

from fuzzywuzzy import fuzz

# Define a similarity threshold (e.g., 70 means 70% similar)
threshold = 70

df_shots['h_a'] = df_shots['teamName'].apply(
    lambda x: 'h' if fuzz.ratio(x.lower(), h_team.lower()) >= threshold else 'a'
)

df_shots.head()
```

```python
# Option for Peñarol
from fuzzywuzzy import fuzz

threshold = 80  # Adjust based on testing

df_shots['h_a'] = df_shots['teamName'].apply(
    lambda x: 'h' if fuzz.partial_ratio(x.lower(), h_team.lower()) >= threshold else 'a'
)

df_shots.head()
```
